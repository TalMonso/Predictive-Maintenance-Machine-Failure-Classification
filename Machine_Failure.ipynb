```python
import pandas as pd
import numpy as np
```

```python
data = {'Data' : ['x=a', 'x=b', 'x=c', 'x=d'],
        'z=i' : [10, 2, 12, 12],
        'z=j' : [18, 10, 8, 17],
        'z=k' : [15, 20, 3, 9]}

df = pd.DataFrame(data, columns=['Data', 'z=i', 'z=j', 'z=k'])
```

```python
df['Total'] = df.iloc[:, 1:].sum(axis=1)

if df.iloc[-1, 0] != 'Total':
    total_row = ['Total'] + df.iloc[:, 1:].sum(axis=0).tolist()
    df.loc[len(df)] = total_row

df
```

```python
#Q1

P_x = np.array([43/136, 32/136, 23/136, 38/136])
entropy_x = -np.sum(P_x * np.log2(P_x))
print(f"Entropy of X: {entropy_x}")
```

```python
#Q2

P_z = np.array([36/136, 53/136, 47/136])
entropy_z = -np.sum(P_z * np.log2(P_z))
print(f"Entropy of Z: {entropy_z}")
```

```python
# Q3
# Max entropy of x for 4 cases to x so 25% probability its 1/4

max_entropyX = -np.log2(1/4)
print(f"Maximal Entropy of X: {max_entropyX}")
```

```python
# Q4
# Max entropy of z for 3 cases its 1/3

max_entropyZ = -np.log2(1/3)
print(f"Maximal Entropy of Z: {max_entropyZ}")
```

```python
# Q5
# Entropy H(z|x)= P(X) * H(Z|X)

p_z_of_x = np.array([10/43,2/32,12/23,12/38,18/43,10/32,8/23,17/38,15/43,20/32,3/23,9/38])
p_z_of_x = -np.log2(p_z_of_x)
px = np.array([10,2,12,12,18,10,8,17,15,20,3,9])
px = px/np.sum(px)
p_z_of_x = px * p_z_of_x
entropy_z_of_x = np.sum(p_z_of_x)
print(f"Entropy H(Z|X): {entropy_z_of_x}")
```

```python
# Q6
# Entropy H(X|Z) = P(Z) * H(X|Z)

p_x_of_z = np.array([10/36,18/53,15/47,2/36,10/53,20/47,12/36,8/53,3/47,12/36,17/53,9/47])
p_x_of_z = -np.log2(p_x_of_z)
pz = np.array([10,18,15,2,10,20,12,8,3,12,17,9])
pz = pz/np.sum(pz)
p_x_of_z = pz * p_x_of_z
entropy_x_of_z = np.sum(p_x_of_z)
print(f"Entropy H(X|Z): {entropy_x_of_z}")
```

```python
# Q7
# Mutual information => I(X;Z) = I(Z;X) = H(Z) - H(Z|X) = P(X.Y) * LOG(P(Y|X)/P(Y))

i_xz = entropy_z - entropy_z_of_x
i_zx = entropy_x - entropy_x_of_z
print(f"I(X;Z) = {i_xz} and I(Z;X) = {i_zx}")
```
